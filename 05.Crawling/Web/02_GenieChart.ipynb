# Genie Top200 Chart

## 1. 인터넷 상에서 데이터 가져오기

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Genie Top200
header = {'User-Agent':
          'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'}   # 나는 기계가 아니라 사람이야!

url = 'https://www.genie.co.kr/chart/top200'
req = requests.get(url, headers = header)
html = req.text
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

## 2. 원하는 데이터의 태그 찾기

```python
# <table class="list-wrap">
table = soup.select_one('table.list-wrap')
trs = table.select('tr.list')
len(trs)
```

## 3. 여러개의 데이터중 하나를 선택해서 원하는 정보 추출

```python
# rank
tr = trs[0]
rank = int(tr.select_one('td.number').get_text().split('\n')[0])
rank
```

```python
# title
title = tr.select_one('.info').select_one('.title').get_text().strip()
title
```

```python
# artist
artist = tr.select_one('.info').select_one('.artist').get_text()
artist
```

```python
# album
album = tr.select_one('.info').select_one('.albumtitle').get_text().strip()
album
```

## 한 페이지에 있는 모든 데이터를 반복문으로 가져오기

```python
rank_list, title_list, artist_list, album_list = [], [], [], []

for tr in trs:
    rank = int(tr.select_one('td.number').get_text().split('\n')[0])
    title = tr.select_one('.info').select_one('.title').get_text().strip()
    artist = tr.select_one('.info').select_one('.artist').get_text()
    album = tr.select_one('.info').select_one('.albumtitle').get_text().strip()

    rank_list.append(rank)
    title_list.append(title)
    artist_list.append(artist)
    album_list.append(album)
```

```python
genie_dict = dict(rank=rank_list, title=title_list, artist=artist_list, album=album_list)
df = pd.DataFrame(genie_dict)
df
```

```python
genie_dict = dict(순위=rank_list, 곡명=title_list, 가수=artist_list, 앨범=album_list)
df = pd.DataFrame(genie_dict)
df
```

### 5. 모든 페이지 데이터 가져오기

```python
sub = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210817&hh=14&rtm=Y&pg='
rank_list, title_list, artist_list, album_list = [], [], [], []

for page in range(1,5):
    url = f'{sub}{page}'
    req = requests.get(url, headers=header)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    trs = soup.select('tr.list')
    for tr in trs:
        rank = int(tr.select_one('td.number').get_text().split('\n')[0])
        title = tr.select_one('.info').select_one('.title').get_text().strip()
        artist = tr.select_one('.info').select_one('.artist').get_text()
        album = tr.select_one('.info').select_one('.albumtitle').get_text().strip()

        rank_list.append(rank)
        title_list.append(title)
        artist_list.append(artist)
        album_list.append(album)
```

```python
genie_dict = dict(순위=rank_list, 곡명=title_list, 가수=artist_list, 앨범=album_list)
df = pd.DataFrame(genie_dict)
df
```
