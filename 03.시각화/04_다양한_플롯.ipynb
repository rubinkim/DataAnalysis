## 1. Bar Chart
- x 데이터가 카테고리 값인 경우

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['A','B','C']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("ABC")
plt.ylabel("Count")
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

## 2.파이 차트

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

### Anscombe's Qurartet
- Bubble chart할 때 반드시 다뤄야 할 것
- 선형회귀선 : y = 3.0 + 0.50 * x

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.info()
```

```python
ans.head()
```

```python
ans.groupby('dataset').describe().round(4)
```

- Dataset1에 대하여 산점도와 회귀선 그리기

```python
ds1 = ans[ans.dataset == 'I']
plt.scatter(x=ds1.x, y=ds1.y)
plt.plot([2, 16], [3.0+0.5*2, 3.0+0.5*16], c='r', lw=2)
plt.grid()
plt.title('Dataset I')
plt.show()
```

```python
np.unique(ans.dataset)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(8,8))
for i, roman in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset ==roman]
    ax = axes[i//2, i%2]
    ax.scatter(ds.x, ds.y)
    ax.plot([3,20],[4.5,13], 'r-', lw=2)
    ax.set_title(f'Dataset {roman}')
    ax.grid()

fig.suptitle("Anscombe's Quartet", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

## 6.Box plot

```python
iris = sns.load_dataset('iris')
iris.head()
```

# Setosa 품종의 각 feature의 박스 플롯

```python
iris.boxplot(column='sepal_length', by='species')
plt.grid('off')
```

```python
fig, axes = plt.subplots(2,2, figsize=(20,10))
for i, feature in enumerate(iris.columns[:-1]):
    ax = axes[i//2, i%2]
    iris.boxplot(column=feature, by='species', ax=ax)
    ax.set_xlabel('')

fig.suptitle(f'Box Plot for features by species', fontsize=20, y=1.0)
plt.tight_layout()
plt.show()
```

- 모든 품종의 피쳐별 박스 플롯

```python
fig, axes = plt.subplots(1,3, figsize=(20, 6), sharey=True)

for i, species in enumerate(iris.species.unique()):
    sub = iris[iris.species == species]
    ax = axes[i]
    ax.boxplot([sub.sepal_length, sub.sepal_width, sub.petal_length, sub.petal_width], labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
    ax.set_title(species)

fig.suptitle("Box plot of Every Species By Features", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
