# Seaborn 데이터 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

## 1. 카운트 플롯

```python
sns.countplot(x='class', data=titanic)
plt.title('Passengers by Class')
plt.show()
```

```python
sns.countplot(x='day', data=tips)
plt.title('Tip by Days')
plt.show()
```

## 2. 다차원 실수형 데이터
- pairplot()

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue='species', markers=list('osD'));
```

## 3. 2차원 카테고리 데이터
- heatmap()

```python
tt = titanic.pivot_table(index='sex' , columns='class', aggfunc='size')
tt
```

```python
sns.heatmap(data=tt, annot=True, fmt='d')
plt.title('Titanic Pclass gender')
plt.show()
```

```python
sns.heatmap(data=tt, annot=True, fmt='d', cmap=sns.light_palette('purple'))
plt.title('Titanic Pclass gender')
plt.show()
```

```python
flights_passengers = flights.pivot("month", "year", "passengers")
plt.title("연도, 월 별 승객수에 대한 Heatmap")
sns.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1)
plt.show()
```
